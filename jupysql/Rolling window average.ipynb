Calculate daily sum of all transaction amounts

```python
%%sql --save daily_transactions  
SELECT CAST(transaction_time as DATE) AS date, SUM(transaction_amount) AS daily_transaction_amount
FROM transactions
GROUP BY date
ORDER BY date
```

```python
%%sql --with daily_transactions  # Query to calculate 3 day rolling window average for each day.
SELECT date, AVG(daily_transaction_amount) 
OVER (ORDER BY date ROWS BETWEEN 2 PRECEDING AND CURRENT ROW) AS rolling_avg
FROM daily_transactions
```
